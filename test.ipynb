```python
"""
WordNet Relation Explorer
Prints every relation available in WordNet for a given word.

Usage:
    python wordnet_explorer.py <word>
    python wordnet_explorer.py <word> --pos n|v|a|r
    python wordnet_explorer.py  (runs interactive mode)

Requirements:
    pip install nltk
    python -c "import nltk; nltk.download('wordnet'); nltk.download('omw-1.4')"
"""

import sys
import textwrap
from collections import defaultdict

try:
    from nltk.corpus import wordnet as wn
except ImportError:
    print("NLTK not installed. Run: pip install nltk")
    sys.exit(1)


POS_LABELS = {
    "n": "noun",
    "v": "verb",
    "a": "adjective",
    "s": "adjective satellite",
    "r": "adverb",
}

SECTION_WIDTH = 72


def hr(char="─"):
    return char * SECTION_WIDTH


def heading(title):
    pad = SECTION_WIDTH - len(title) - 2
    left = pad // 2
    right = pad - left
    return f"{'─' * left} {title} {'─' * right}"


def fmt_synset(s):
    """Short display name for a synset: word (pos) — definition snippet."""
    word = s.lemma_names()[0].replace("_", " ")
    pos = POS_LABELS.get(s.pos(), s.pos())
    defn = s.definition()
    snippet = (defn[:55] + "…") if len(defn) > 58 else defn
    return f"{word}  [{pos}]  — {snippet}"


def fmt_lemma(l):
    return l.name().replace("_", " ")


def wrap(items, indent=4, width=SECTION_WIDTH):
    if not items:
        return " " * indent + "(none)"
    bullet = " " * indent + "• "
    lines = []
    for item in items:
        wrapped = textwrap.fill(
            item,
            width=width,
            initial_indent=bullet,
            subsequent_indent=" " * (indent + 2),
        )
        lines.append(wrapped)
    return "\n".join(lines)


def collect_synset_relations(synset):
    """
    Return a dict of relation_name -> list of display strings for a synset.
    Covers every relation available at the synset level.
    """
    r = defaultdict(list)

    for s in synset.hypernyms():
        r["hypernyms (broader / is-a)"].append(fmt_synset(s))
    for s in synset.instance_hypernyms():
        r["instance hypernyms (this is an instance of)"].append(fmt_synset(s))
    for s in synset.hyponyms():
        r["hyponyms (narrower / kinds of this)"].append(fmt_synset(s))
    for s in synset.instance_hyponyms():
        r["instance hyponyms (specific instances)"].append(fmt_synset(s))
    for s in synset.similar_tos():
        r["similar to (adj cluster members)"].append(fmt_synset(s))
    for s in synset.also_sees():
        r["also see (related adjectives)"].append(fmt_synset(s))
    for s in synset.attributes():
        r["attributes (nouns this adj describes)"].append(fmt_synset(s))
    for s in synset.entailments():
        r["entailments (verb implies)"].append(fmt_synset(s))
    for s in synset.causes():
        r["causes (this verb causes)"].append(fmt_synset(s))
    for s in synset.verb_groups():
        r["verb groups (related verb senses)"].append(fmt_synset(s))
    for s in synset.part_meronyms():
        r["part meronyms (parts of this)"].append(fmt_synset(s))
    for s in synset.part_holonyms():
        r["part holonyms (wholes that include this)"].append(fmt_synset(s))
    for s in synset.member_meronyms():
        r["member meronyms (members of this group)"].append(fmt_synset(s))
    for s in synset.member_holonyms():
        r["member holonyms (groups this belongs to)"].append(fmt_synset(s))
    for s in synset.substance_meronyms():
        r["substance meronyms (substances in this)"].append(fmt_synset(s))
    for s in synset.substance_holonyms():
        r["substance holonyms (things made of this)"].append(fmt_synset(s))
    for s in synset.topic_domains():
        r["topic domain"].append(fmt_synset(s))
    for s in synset.region_domains():
        r["region domain"].append(fmt_synset(s))
    for s in synset.usage_domains():
        r["usage domain"].append(fmt_synset(s))
    for s in synset.in_topic_domains():
        r["in topic domain (synsets in this topic)"].append(fmt_synset(s))
    for s in synset.in_region_domains():
        r["in region domain"].append(fmt_synset(s))
    for s in synset.in_usage_domains():
        r["in usage domain"].append(fmt_synset(s))

    return r


def collect_lemma_relations(lemma):
    """
    Return a dict of relation_name -> list of display strings for a lemma.
    Covers every relation available at the lemma level.
    """
    r = defaultdict(list)

    for l in lemma.antonyms():
        r["antonyms"].append(fmt_lemma(l))
    for l in lemma.derivationally_related_forms():
        r["derivationally related forms"].append(
            f"{fmt_lemma(l)}  [{POS_LABELS.get(l.synset().pos(), l.synset().pos())}]"
        )
    for l in lemma.pertainyms():
        r["pertainyms (adj → noun it pertains to)"].append(fmt_lemma(l))
    for l in lemma.also_sees():
        r["also see"].append(fmt_lemma(l))

    return r


def print_synset_block(synset, word):
    """Print a full block of output for one synset."""
    pos_label = POS_LABELS.get(synset.pos(), synset.pos())
    synonyms = [
        n.replace("_", " ")
        for n in synset.lemma_names()
        if n.lower() != word.lower()
    ]
    examples = synset.examples()

    print(heading(f"{synset.name()}  [{pos_label}]"))
    print()
    print(f"  Definition : {synset.definition()}")
    if synonyms:
        print(f"  Synonyms   : {', '.join(synonyms)}")
    else:
        print(f"  Synonyms   : (this word is the only lemma in this synset)")
    if examples:
        for ex in examples:
            print(f'  Example    : "{ex}"')
    print()

    synset_rels = collect_synset_relations(synset)

    print("  ── Synset-level relations ──────────────────────────────────────")
    if synset_rels:
        for rel_name, entries in synset_rels.items():
            print(f"\n  {rel_name.upper()}")
            print(wrap(entries, indent=4))
    else:
        print("    (no synset-level relations for this sense)")

    print()
    print("  ── Lemma-level relations ───────────────────────────────────────")
    any_lemma_rel = False
    for lemma in synset.lemmas():
        lemma_rels = collect_lemma_relations(lemma)
        if lemma_rels:
            any_lemma_rel = True
            print(f"\n  Lemma: {fmt_lemma(lemma)}")
            for rel_name, entries in lemma_rels.items():
                print(f"    {rel_name.upper()}")
                print(wrap(entries, indent=6))
    if not any_lemma_rel:
        print("    (no lemma-level relations for this sense)")

    print()


def explore(word, pos_filter=None):
    synsets = wn.synsets(word, pos=pos_filter)

    if not synsets:
        msg = f"No WordNet entries found for '{word}'"
        if pos_filter:
            msg += f" with POS filter '{pos_filter}'"
        print(msg)
        return

    print()
    print(hr("═"))
    header = f"  WordNet relations for: {word.upper()}"
    if pos_filter:
        header += f"  (pos={pos_filter})"
    print(header)
    print(f"  {len(synsets)} sense{'s' if len(synsets) != 1 else ''} found")
    print(hr("═"))
    print()

    for synset in synsets:
        print_synset_block(synset, word)
        print()


def interactive_mode():
    print()
    print("WordNet Relation Explorer")
    print("Type a word to explore, or 'quit' to exit.")
    print("Optional: add POS filter, e.g.  happy --pos a")
    print()

    while True:
        try:
            raw = input("word> ").strip()
        except (EOFError, KeyboardInterrupt):
            print()
            break

        if not raw or raw.lower() in ("quit", "exit", "q"):
            break

        parts = raw.split()
        word = parts[0]
        pos_filter = None

        if "--pos" in parts:
            idx = parts.index("--pos")
            if idx + 1 < len(parts):
                pos_filter = parts[idx + 1]

        explore(word, pos_filter)


def main():
    args = sys.argv[1:]

    if not args:
        interactive_mode()
        return

    word = args[0]
    pos_filter = None

    if "--pos" in args:
        idx = args.index("--pos")
        if idx + 1 < len(args):
            pos_filter = args[idx + 1]

    explore(word, pos_filter)


if __name__ == "__main__":
    main()
```

```python
import wn

en = wn.Wordnet('oewn:2024')

word = en.words('happy')[0]
for sense in word.senses():
    synset = sense.synset()
    print(f"\nSense: {sense.id}")
    
    similar = synset.get_related('similar')
    also    = synset.get_related('also')
    
    print("  similar:", [s.lemmas() for s in similar])
    print("  also:",    [s.lemmas() for s in also])
```

```python
def get_synset_relation(word_str, relation):
    results = set()
    for word in en.words(word_str):
        for sense in word.senses():
            for related_synset in sense.synset().get_related(relation):
                for lemma in related_synset.lemmas():
                    results.add(lemma)
    return results

print(get_synset_relation('happy', 'similar'))
print(get_synset_relation('happy', 'also'))
```
